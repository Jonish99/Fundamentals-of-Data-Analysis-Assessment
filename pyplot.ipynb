```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-2.0*np.pi,2.0*np.pi,0.1)

plt.plot(x,np.sin(x),'g.')
plt.plot(x,np.cos(x),'b.')

plt.show()
```

```python
plt.plot([1, 2, 3, 4],[1,4,9,16],'b.')
plt.ylabel('some numbers')
plt.show()
```

```python
x = np.arange(0.0,10.0,0.01)
y = 3.0 * x +1.0

noise = np.random.normal(0.10,1.0,len(x))

plt.plot(x,y + noise,'c.', label="noise" )
plt.plot(x,y,'g.', label="strt" )
plt.title("Simple plot")
plt.xlabel('weight')
plt.ylabel('mass')
plt.legend()

plt.show()
```

```python
x = np.arange(0.0,10.0,0.01)
y = 3.0 * x +1.0

noise = np.random.normal(0.10,1.0,len(x))
#repeat calls to the plot
plt.plot(x,y + noise,'r.' )
plt.plot(x,y,'b.')

plt.show()
#close plot, then new plot
plt.plot(x,y + noise,'c.' )
plt.plot(x,y,'g.')

plt.show()
```

```python

x=np.random.normal(0.0,1.0,1000)

#histogram plot creates bins n the range
plt.hist(x, bins =25)
plt.show()
```

```python
x = np.arange(1.0,10.0,0.1)

plt.plot(x, x**2,'g-',label="x^2")
plt.plot(x, x**3,'b-',label="x^3")
plt.plot(x, x**4,'r-',label="x^4")
plt.plot(x, x**5,'k-',label="x^x")

plt.legend()
plt.show()
```

```python
x = np.random.uniform(0.0,10.0,100)
y = np.random.uniform(0.0,100.0,100)
z = np.random.normal(100.0,40.0,100)
c = np.random.randint(0,100,100)

plt.scatter(x,y,c=c,s=z)

plt.show()
```

```python

#one row two cols
plt.subplot(1,2,1)

x=np.random.normal(0.0,1.0,1000)
plt.hist(x)

plt.subplot(1,2,2)
x=np.random.uniform(-3.0,3.0,1000)
plt.hist(x)


plt.show()
```
