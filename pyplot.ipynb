Graphs in Matplotlib are drawn on *figures* which can contain one or more axes. 
The figure acts as a *container* for 'child' axes which includes *artists* such as the title, legend and the canvas (not too relevant to matplot users)(Usage Guide â€” Matplotlib 3.5.0 documentation, 2021). In pyplot, axes refers to the axis which create a plot space usually two but sometimes three in 3D plots, i.e a plot. A figure will contain at least one axes but may also contain multiple axes. A subplot is one of the axes on a mulit-axes figure. Is configured with a tuple defining the grid layout. For example, subplot(2,3) will produce a grid with 3 rows each able to contain a plot. The location of an axes on any position in the grid is controled via an array value passed to the grid, for example, ax[0,2].plot(x) will place the axes in the 1st row from the top and 3rd column from the left of the grid.

Axis are the lines usually seen as the y and x axis which numbers are shown, delimited by the upper and lower boundaries of the array passed. The following plot is adapted from a range of soureces to demonstrate many of the figure properties


```python
# (How To Display A Plot In Python using Matplotlib - ActiveState, 2021)

fig, ax = plt.subplots()
line1, = ax.plot([1,2,3], label="Line 1", linestyle='--')
line2, = ax.plot([3, 2, 1], label="Line 2", linewidth=4)

# Create a legend. The legend is reference to line on the plotas the axes in added to the variable line1 above. handles=[line1]
legend1 = ax.legend(handles=[line1], loc='upper right')
#Legend a can be added manually, eg Legend 1
# Add the legend manually to the Axes.
ax.add_artist(legend1)
#Or as in the case of this legend added directly to the axes.
ax.legend(handles=[line2], loc='lower right')
#need to add ticks, titles, legend handlers and more
plt.rcParams["figure.figsize"] = [7.50, 3.50]#Users can also control the size of the figure
#rather than show all the ticks on an axis the MaxLocator method allows users to control the number of ticks on any axis
#(VanderPlas, 2021)
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) 
ax.yaxis.set_major_locator(plt.MaxNLocator(6)) 

plt.show()
```

In this example adapted (VanderPlas, 2021) from  we look at how we can create tick labels

```python
# Plot a sine and cosine curve
fig, ax = plt.subplots()
plt.rcParams["figure.figsize"] = [7.50, 3.50]
x = np.linspace(0, 3 * np.pi, 1000) # create a sequence of 1000 numbers between 0 and 3 * pi for the x axis
#plot 1
ax.plot(x, np.sin(x), lw=3, label='Sine') # use the sine of x for the y axis on subplot
#plot 2
ax.plot(x, np.cos(x), lw=3, label='Cosine')# use consine of x for the y axis on subplot.

# Set up grid, legend, and limits
#plt.rcParams['legend.fontsize'] = 10
ax.grid(True) #create a grid
ax.legend(frameon=False)
ax.axis('equal')#
ax.set_xlim(0, 3 * np.pi)# ensure the x axis starts at 0 and finishes at 3 * pi, 
ax.set_ylim(0, 2)
#in this case it means the lines extend to the edge of the plot, i.e the x axis has the same range as the values on x.

#set the major and minor tics at user determined values
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))#half values of pi
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))#quarter values of pi

#The following code is from VanderPlas, 2021

#Now to custom label the ticks.
#Use  Matplotlib's LaTeX support - allowing the printing of mathetmatical symbols, i.e. pi
def format_func(value, tick_number):
    # find values on the pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0" #1st value is 0
    elif N == 1:
        return r"$\pi/2$" #2nd value is pi/2
    elif N == 2:
        return r"$\pi$" #3rd value is pi
    elif N % 2 > 0: # for the remaining values on the odd value-3, (even position -4) e.g 3*pi/2
        return r"${0}\pi/2$".format(N) 
    else:
        return r"${0}\pi$".format(N // 2) #the even value * pi
#use the function to set the ticks labels.
#The plt funcformatter takes a function to create the tick labels

ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
```

The following examples demonstrates the use of multiple plots(axes) on a figure. Individual plots are known as a subplots.

```python
# Create a Figure with 2 rows and 2 columns of subplots: 
#using axes
#adapted from: (How To Display A Plot In Python using Matplotlib - ActiveState, 2021)
fig, ax = plt.subplots(2, 2)
#linspace created sequence of 100 evenly spaced number between 0 and 100
x = np.linspace(0, 5, 100)
#print (x)
# Index 4 Axes arrays in 4 subplots within 1 Figure: 
ax[0, 0].plot(x, np.sin(x), 'g') #row=0, column=0
ax[1, 0].plot(range(200), 'b') #row=1, column=0

ax[0, 1].plot(x, np.cos(x), 'r') #row=0, column=1
ax[0, 1].plot(x, np.sin(x), 'b') #plot second line on row=0, column=1 
ax[1, 1].plot(x, np.tan(x), 'c') #row=1, column=1

#Adding titles (Matplotlib and Shinji, 2021)
ax[0, 0].title.set_text('Sine')
ax[1, 0].title.set_text('Proportion')
ax[0, 1].title.set_text('Sine & Cosine')
ax[1, 1].title.set_text('Tangent')
fig.subplots_adjust(hspace=.5) # create some padding between plots as titles cover x axis labels 
#(Need to add space between SubPlots for X axis label and Lefevre, 2021)
fig.suptitle("Demonstration of subplots")#set title for whole figure
plt.rcParams["figure.figsize"] = [3.50, 7.50] # set figure size
plt.show()
```

So far we have looked at 2 dimenional plots, usually involving the relationship between two variables. However, the mpl_toolkits.mplot3d library which now ships with matplotlib, can enable the plotting of the relationship between 3 variables.

```python
#(An easy introduction to 3D plotting with Matplotlib, 2021)


fig = plt.figure()
fig.suptitle("Demonstration of 3d plots")#set title
ax = plt.axes(projection="3d") # The projection keyword value ='3d' means from herein the plot is a 3d plot and
#will expect paramters to treat it as such
#create a the plot without any data points
plt.rcParams["figure.figsize"] = [3.50, 3.50]

plt.show()
```

```python
#(mplot3d tutorial â€” Matplotlib 2.0.2 documentation, 2021)
fig = plt.figure()
ax = plt.axes(projection="3d")
fig.suptitle("Demonstration of 3d plots")#set title

#plot a line on the 3 dplot
z_line = np.linspace(0, 18, 1000)#use np.linspace to create 1000 points unformly space betwee 0 * 15
x_line = np.cos(z_line)
y_line = np.sin(z_line)
# A line is plotted using points of 3 values, x,y & z
ax.plot3D(x_line, y_line, z_line, 'gray')
#print(x_line)
#Using the cosinse and sine again on the x and y xis.
#plot scatter points on the 3d plot
z_points = 18 * np.random.random(100)
x_points = np.cos(z_points) + 0.1 * np.random.randn(100)
y_points = np.sin(z_points) + 0.1 * np.random.randn(100)
ax.scatter3D(x_points, y_points, z_points, c=z_points, cmap='hsv');
plt.rcParams["figure.figsize"] = [7.50, 7.50] #set figure size
#label the axis
xLabel = ax.set_xlabel('X axis')
yLabel = ax.set_ylabel('Y axis')
zLabel = ax.set_zlabel('Z axis')
plt.show()
```

#### 2. Seaborn Pairplot
The Seaborn pair plot creates a grid of plot such that each plot in a single column of the figure will share the same x axis and each plot in a single row will share the same y axis.

In this example we will use the nba.csv data set of US National Basekball Association. (Data-Analysis-with-Pandas-and-Python/nba.csv at master Â· sivabalanb/Data-Analysis-with-Pandas-and-Python, 2021)

(Data visualization with Pairplot Seaborn and Pandas - GeeksforGeeks, 2021)

```python
# load the nba dataset into pandas dataframe

df = pd.read_csv("data/nba.csv")

#the height format is in feet and inches. e.g. 7-2
#Use function to convert that column to inches(pandas and Horvitz, 2021)

def parse_ht(ht):
    # format: 7' 0.0"
    #split an on '-' and create array
    ht=str(ht)
    ht_ = ht.split(".")
    #print(ht_[1])
    #convert feet to float
    ft_ = float(ht_[1])
    #repalce double quote with nothing
    in_ = float(ht_[1].replace("\"",""))
    #print((12*ft_) + in_)
    #feet * 12 + inches gives total inches
    return int((12*ft_) + in_)
df.dropna()
#apply function to column
df["Height"] = df["Height"].apply(lambda x:parse_ht(x))
df.dropna()

#specify variable to use from dataframe
sns.pairplot(df,vars=["Height","Age","Weight","Salary"], diag_kind='hist', height=1.4)
```

The pairplot takes a range of parameters. 
vars will allow control of specific column varialbles to be shown. e.g. vars=["Height","Age"] will produce a 2 x 2 plot.
*hue* will allow users to take a variable and control the shade of each datapoint based on a variable value, e.g. sns.pairplot(df.head(), hue = 'Age')

The default plot is a scatter but the king parameter allows the users some control. kind = {â€˜scatterâ€™, â€˜kdeâ€™, â€˜histâ€™, â€˜regâ€™}

diag_kind{â€˜autoâ€™, â€˜histâ€™, â€˜kdeâ€™, None} - type of diagonal plot, where the plot is univariate, e.g. 

markers=["o", "s", "D"] can be used but in the example below species is the hue and different markers do not add to the plot.


```python
#The following pairplot has code has been adapted from (seaborn.pairplot â€” seaborn 0.11.2 documentation, 2021)
penguins = sns.load_dataset("penguins")
g=sns.pairplot(penguins, hue="species",kind='kde',height=1.4) 
# use penguin species for hue and Kernel Density Estimation for plot type and histograms for the diagonal type. The height setting is for the subplot

g.fig.suptitle("Pairplot illustrating penguin species variables", y=1.08) # set title and height for title
plt.show()
```

#### 3. Ridgeline Graph  (Python-graph-gallery.com. 2021.)
The following example has been adapted from (Python-graph-gallery.com. 2021.), but rather than use the same data set, this example has been reconfigured to show Cork mean monthly rainfall from 1941 to 1994.(Cork (Univ.Coll.) Climate Data - Cork (Univ.Coll.) Climate Data - data.gov.ie, 2021).

The data set has been manually *cleaned*. The mean monthly rainfall for each month is calculated for the years in scope. 
A seaborn Facetgrid is used and a kde plot for each month is plotted for each month. The density of each month indicates the amount of rain. 

```python

sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

# getting the data
rain = pd.read_csv('data/dly904.csv') # we retrieve the data from plotly's GitHub repository

rain['month'] = pd.to_datetime(rain['date']).dt.month # we store the month in a separate column
rain['year'] = pd.to_datetime(rain['date']).dt.year # we store the month in a separate column
# Define dictionary to apply to dataframe column to use a labels when the plots are created.
month_dict = {1: 'January',
              2: 'February',
              3: 'March',
              4: 'April',
              5: 'May',
              6: 'June',
              7: 'July',
              8: 'August',
              9: 'September',
              10: 'October',
              11: 'November',
              12: 'December'}

# Create a new column, 'month'
rain['month'] = rain['month'].map(month_dict)
# Create a pandas serie calculating the mean monthly rainfall for years in scope. 
month_mean_serie = rain.groupby('month')['daily_rain'].mean()
year_mean_serie = rain.groupby('year')['daily_rain'].sum()
#print(month_mean_serie)
#Apply mean monthyly rainfall to a new column mapped on months.
rain['mean_month'] = rain['month'].map(month_mean_serie)
rain['mean_year'] = rain['month'].map(year_mean_serie)

rain = rain.groupby(['month','year'])['daily_rain'].sum()
print(rain)
```

```python
# Generate a 12 color seaborn pallette, one for each month

pal = sns.color_palette(palette='BuPu_r', n_colors=12)

# in the sns.FacetGrid class, the 'hue' argument is the one that is the one that will be represented by colors with 'palette'
g = sns.FacetGrid(rain, row='month', hue='mean_month', aspect=15, height=0.75, palette=pal)

# then we add the densities kdeplots for each month
g.map(sns.kdeplot, 'daily_rain',
      bw_adjust=1, clip_on=False,
      fill=True, alpha=1, linewidth=1.5)

# here we add a white line that represents the contour of each kdeplot
g.map(sns.kdeplot, 'daily_rain', 
      bw_adjust=1, clip_on=False, 
      color="w", lw=2)

# here we add a horizontal line for each plot
g.map(plt.axhline, y=0,
      lw=2, clip_on=False)

# we loop over the FacetGrid figure axes (g.axes.flat) and add the month as text with the right color
# notice how ax.lines[-1].get_color() enables you to access the last line's color in each matplotlib.Axes
for i, ax in enumerate(g.axes.flat):
    ax.text(-15, 0.05, month_dict[i+1],
            fontweight='bold', fontsize=15,
            color=ax.lines[-1].get_color())
    
# we use matplotlib.Figure.subplots_adjust() function to get the subplots to overlap
g.fig.subplots_adjust(hspace=-0.3)

# eventually we remove axes titles, yticks and spines
g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True)

plt.setp(ax.get_xticklabels(), fontsize=15, fontweight='bold')
plt.xlabel('MM rain', fontweight='bold', fontsize=15)
g.fig.suptitle('Monthly Average Rainfall in Cork',
               ha='right',
               fontsize=20,
               fontweight=20)

plt.show()
```
